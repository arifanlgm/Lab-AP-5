Disini saya memvisualisasikan kategori kualitas udara dengan menggunakan style dark horizontal ini

```python
from matplotlib import pyplot as plt

import seaborn as sns

df.groupby('AQI category').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```
